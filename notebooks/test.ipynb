```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

```

```python
# Из папки notebooks выходим в корень и идем в data
# 1. Загрузка данных

purchases = pd.read_csv('../data/raw/apparel-purchases.csv')
messages = pd.read_csv('../data/raw/apparel-messages.csv')
target = pd.read_csv('../data/raw/apparel-target_binary.csv')
```

```python
# 2. Агрегация покупок
client_features = purchases.groupby('client_id').agg({
    'price': ['sum', 'mean'],
    'quantity': 'sum',
    'date': 'max'
})

# Схлопываем MultiIndex: преобразуем ('price', 'sum') в 'price_sum'
client_features.columns = [
    '_'.join(col).strip() if isinstance(col, tuple) else col 
    for col in client_features.columns.values
]

# Сбрасываем индекс, чтобы client_id стал обычной колонкой
client_features = client_features.reset_index()

# 3. Теперь объединение пройдет гладко
data = client_features.merge(target, on='client_id', how='left')
```
